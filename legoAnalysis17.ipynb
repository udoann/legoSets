### What percentage of all licensed sets ever released were Star Wars themed?

```python
merged_data = df.merge(theme, left_on='parent_theme', right_on='name')
merged_data.head()
```

```python
merged_data.drop(columns='name_y', inplace= True)
```

```python
merged_data.head()
```

```python
merged_data[merged_data['set_num'].isnull()].shape
```

```python
# no == sign coz the is_licensed column is a boolean value
licensed = merged_data[merged_data['is_licensed']]
licensed = licensed.dropna(subset=["set_num"])
licensed
```

```python
star_wars = merged_data[merged_data['parent_theme']== 'Star Wars']
star_wars.head()
```

```python
percentage = int((star_wars.shape[0]/licensed.shape[0])*100)
percentage
```

### In which year was star was not the most popular licensed theme(in terms of number of sets released that year) 
##### save your answer as a variable new _era as in the form of an integer

```python
sort_year = star_wars.groupby('year').count().sort_values(by='set_num', ascending= True)
sort_year  # Reset_index adds index
```

```python
maxi_star = sort_year['set_num'].min()
maxi_star
new_era = int(maxi_star)
new_era
```

```python
licensed_num = licensed.groupby(['year','parent_theme']).sum().reset_index()
licensed_num
```

```python
lice_num = licensed_num[licensed_num['parent_theme'] == 'Star Wars']
lice_num.sort_values(by='is_licensed', ascending= True)
```

```python
licensed_sorted = licensed.sort_values('year')
licensed_sum = licensed_sorted.groupby(['year','parent_theme']).sum().reset_index()  # create a count column = 1 but ince the is_licensed column is a boolean we use it to sum
licensed_sum
```

```python
max_licensed = licensed_sum.sort_values('is_licensed',ascending=False).drop_duplicates(['year'])
max_licensed
```
